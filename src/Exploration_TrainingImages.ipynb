```python
from datetime import timedelta, datetime

def get_bbox(fire):
#     xmin = fire['minx'] - (fire['maxx']-fire['minx'])/2
#     xmax = fire['maxx'] + (fire['maxx']-fire['minx'])/2    
#     ymin = fire['miny'] - (fire['maxy']-fire['miny'])/2
#     ymax = fire['maxy'] + (fire['maxy']-fire['miny'])/2
#     return [xmin, ymin, xmax, ymax]
    return fire['buffer'].bounds

def get_time_interval(fire, t_pre=20, t_post=0):
    datestr_f = fire['FireDiscoveryDateTime'][:10] 
    date_f = datetime.strptime(datestr_f, "%Y-%m-%d")
    datestr_f = fire['FireDiscoveryDateTime'][:10] 
    date_f = datetime.strptime(datestr_f, "%Y-%m-%d")
    delta_pre = timedelta(days=t_pre)
    delta_post = timedelta(days=t_post)
    date_i = date_f - delta_pre #+ timedelta(days=8)
    datestr_i = datetime.strftime(date_i, "%Y-%m-%d")
    date_f = date_f + delta_post #+ timedelta(days=8)
    datestr_f = datetime.strftime(date_f, "%Y-%m-%d")

    return [datestr_i, datestr_f]

evalscript_true_color = """
    //VERSION=3

    function setup() {
        return {
            input: [{
                bands: ["B02", "B03", "B04"]
            }],
            output: {
                bands: 3
            }
        };
    }

    function evaluatePixel(sample) {
        return [sample.B04, sample.B03, sample.B02];
    }
"""

def get_first_full_image(data, unique_acquisitions, results_cloud_covers):
    for image, timestamp, cc in zip(data, unique_acquisitions, results_cloud_covers):
        if cc<0.1 and len(image[image==0]) < 0.1 * len(image[image>0]):
            return 'Success', timestamp, cc
    return 'No non-zero images in timeframe', None, None

def get_valid_timestamp(fire_bbox, fire_row, max_days_pre=20):
    results = []
    unique_acquisitions = None
    fire_time_interval = get_time_interval(fire_row, t_pre=14, t_post=-5)
    # Find valid dates in catalog
    search_iterator = catalog.search(
        DataCollection.SENTINEL2_L2A,
        bbox=fire_bbox,
        time=fire_time_interval,
        query={
            "eo:cloud_cover": {
                "lt": 5
            }
        },
        fields={
            "include": [
                "id",
                "properties.datetime",
                "properties.eo:cloud_cover"
            ],
            "exclude": []
        }
    )
    results = list(search_iterator)
    results_cloud_covers = [r['properties']['eo:cloud_cover'] for r in results] if results else None
    if results and [cc for cc in results_cloud_covers if cc < 0.1]:
        process_requests = []
        time_difference = dt.timedelta(seconds=0)
        all_timestamps = search_iterator.get_timestamps()
        unique_acquisitions = all_timestamps
        for timestamp in unique_acquisitions:
            request = SentinelHubRequest(
                evalscript=evalscript_true_color,
                input_data=[
                    SentinelHubRequest.input_data(
                        data_collection=DataCollection.SENTINEL2_L2A,
                        time_interval=(timestamp - time_difference, timestamp + time_difference)
                    )
                ],
                responses=[
                    SentinelHubRequest.output_response('default', MimeType.PNG)
                ],
                bbox=fire_bbox,
                size=bbox_to_dimensions(fire_bbox, 100),
                config=config
            )
            process_requests.append(request)

        download_requests = [request.download_list[0] for request in process_requests]
        data = client.download(download_requests)
        err_code, timestamp, cc = get_first_full_image(data, unique_acquisitions, results_cloud_covers)
        return err_code, timestamp, cc
    err_code = 'No results in timeframe' if not results else 'All results in timeframe too cloudy'
    return err_code, None, None if not results else results_cloud_covers
```

```python
def index(a, b):
    return ((a - b) / (a + b))
```

## Export true color, NDVI, NDWI, and cloud cover images 

```python
evalscript_true_color = """
    //VERSION=3

    function setup() {
        return {
            input: [{
                bands: ["B02", "B03", "B04"]
            }],
            output: {
                bands: 3
            }
        };
    }

    function evaluatePixel(sample) {
        return [sample.B04, sample.B03, sample.B02];
    }
"""

evalscript_indices = """
    //VERSION=3
    function setup() {
      return {
        input: ["B04", "B08", "B8A", "B11"],
        output: { bands: 4 }
      }
    }

    function evaluatePixel(sample) {
      return [sample.B04, sample.B08, sample.B8A, sample.B11];
    }
"""

evalscript_clm = """
    //VERSION=3
    function setup() {
      return {
        input: ["CLM"],
        output: { bands: 1 }
      }
    }

    function evaluatePixel(sample) {
      if (sample.CLM == 1) {
        return [0.7]
      }
      return [0.1];
    }
"""

def get_image_data(unique_acquisitions, time_difference, evalscript):
    process_requests = []
    for timestamp in unique_acquisitions:
        request = SentinelHubRequest(
            evalscript=evalscript,
            input_data=[
                SentinelHubRequest.input_data(
                    data_collection=DataCollection.SENTINEL2_L2A,
                    time_interval=(timestamp - time_difference, timestamp + time_difference)
                )
            ],
            responses=[
                SentinelHubRequest.output_response('default', MimeType.PNG)
            ],
            bbox=fire_bbox,
            size=bbox_to_dimensions(fire_bbox, 100),
            config=config
        )
        process_requests.append(request)
    download_requests = [request.download_list[0] for request in process_requests]
    data = client.download(download_requests)
    return data[0]
```
